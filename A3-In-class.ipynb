# Prepare data

```python
from pathlib import Path
import re

import numpy
import eelbrain
from matplotlib import pyplot
import mne


# Define some paths that will be used throughout
DATA_ROOT = Path("~").expanduser() / 'Data' / 'Alice'
STIMULUS_DIR = DATA_ROOT / 'stimuli'
EEG_DIR = DATA_ROOT / 'eeg'

# Load one subject's raw EEG file
SUBJECT, ROI = 'S20', ['49', '50', '10', '11']
LOW_FREQUENCY = 0.5
HIGH_FREQUENCY = 20

raw = mne.io.read_raw(EEG_DIR / SUBJECT / f'{SUBJECT}_alice-raw.fif', preload=True)
# Filter the raw data to the desired band
raw.filter(LOW_FREQUENCY, HIGH_FREQUENCY, n_jobs=1)
# Interpolate bad channels
# This is not structly necessary for a single subject. 
# However, when processing multiple subjects, it will allow comparing results across all sensors.
raw.interpolate_bads()
# Load the events embedded in the raw file as eelbrain.Dataset, a type of object that represents a data-table
events = eelbrain.load.mne.events(raw)
```

```python
# Generate the acoustic envelope for all trials in this dataset
envelopes = []
for stimulus_id in events['event']:
    wav = eelbrain.load.wav(STIMULUS_DIR / f'{stimulus_id}.wav')
    envelope = wav.envelope()
    envelope = eelbrain.filter_data(envelope, LOW_FREQUENCY, HIGH_FREQUENCY, pad='reflect')
    envelope = eelbrain.resample(envelope, 100)
    envelopes.append(envelope)
# Add the envelopes to the events table
events['envelope'] = envelopes
# Add a second predictor corresponding to acoustic onsets 
events['onsets'] = [envelope.diff('time').clip(0) for envelope in envelopes]

# Extract the stimulus duration (in seconds) from the envelopes
events['duration'] = eelbrain.Var([envelope.time.tstop for envelope in events['envelope']])

# extract EEG data corresponding exactly to the timing of the envelopes
events['eeg'] = eelbrain.load.mne.variable_length_epochs(events, 0, tstop='duration', decim=5, connectivity='auto')
```

```python
# Get time series for EEG and acoustic onsets
eeg = eelbrain.concatenate(events['eeg']).mean(sensor=ROI)
onsets = eelbrain.concatenate(events['onsets'])
```

# Get array data

```python
# Get stimulus (s) and response (r) arrays for estimation
s = onsets.x
r = eeg.x
# Normalize s and r
r = r - r.mean()
r /= r.std()
s = s - s.mean()
s /= s.std()
```

```python
# Plot the first 5 seconds
pyplot.plot(r[:500])
pyplot.plot(s[:500])
numpy.corrcoef(r, s)[0, 1]
```

# Coordinate descent

```python
from scipy.signal import convolve


def calculate_loss(s, r, h):
    """
    Returns the loss when predicting r with s*h
    """
    r_pred = convolve(s, h, 'full')
    # In full mode, assuming τ starts at 0, r_pred[0] corresponds to s[0]  
    error = r - r_pred[:len(r)]
    error **= 2
    return error.sum()


def calculate_r(s, r, h):
    """
    Returns the correllation between r and predicted r
    """
    r_pred = convolve(s, h, 'full')
    return numpy.corrcoef(r, r_pred[:len(r)])[0, 1]


def find_new_h(s, r, h_curr, delta=0.005):
    """
    Tries to improve h by changing one element of h_curr either by +delta or by -delta.
    Returns the best h and corresponding loss if h can be improved, or None if no improvement can be found.
    """
    curr_loss = calculate_loss(s,r,h_curr)
    best_h = None
    best_loss = numpy.inf
    for delta in (delta, -delta):
        for i in range(len(h_curr)):
            h = h_curr.copy()
            h[i] += delta
            loss = calculate_loss(s,r,h)
            if loss < best_loss:
                best_h = h
                best_loss = loss
    if best_loss <= curr_loss:
        return best_h, best_loss
```

## Basic coordinate descent
- Initialize an all-0 TRF `h`
    - Use length 50 (corresponding to time lags from 0 to 500 ms)
- Use all data to train `h`
- Evaluate the correlation between `r` and `r_pred`, and plot h

```python
curr_h = numpy.zeros(50)

for i in range(999):
    update = find_new_h(s,r,curr_h)
    if update is None:
        curr_h, curr_loss = update

pyplot.plot(curr_h)
```

## Early Stopping
- Reserve the first 1/5 of the data as testing data
    - Reserve the second 1/5 of the data as validation data
    - Use the rest as training data
- Train `h`:
    - Find coordinate descent step using the trainig data
    - After every step, check the error on the validation data
    - When the error on the validation data starts increasing, stop training
- Evaluate `h` on training and testing data
