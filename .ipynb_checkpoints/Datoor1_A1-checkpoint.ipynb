```python
# imports and checking of MNE version

import mne
import numpy as np
import matplotlib.pyplot as plt
import os
import pathlib

print(f"{mne.__version__=}")
```

<h2>1.1 Load/Inspect Data</h2>

```python
data_path = mne.datasets.sample.data_path()
print(data_path)
```

```python
raw_data_file = os.path.join(
    data_path, "MEG", "sample", "sample_audvis_raw.fif"
)

raw_data = mne.io.read_raw_fif(raw_data_file, preload = True)
raw_data.plot()
```

<h2>1.2 Preprocessing </h2>

```python
# Filter signals outside the range of 1-40 hz
filtered_data = raw_data.copy().filter(1, 40)
filtered_data.plot()
```

```python
from mne.preprocessing import ICA

# use 100 components because computation is taking too long, set random state so result stays same
# 100 components takes my computer between 2-3 minutes
# I have changed this value to 10 because it gave the same results
ica = ICA(n_components=10, random_state=10, method = 'fastica')
ica.fit(filtered_data)
```

```python
# plot to check which one is heartbeat
#ica.plot_sources(filtered_data)
```

```python
#From the above plots my guess is that ICA001 is the heartbeat
ica.exclude = [1]
no_heartbeat = ica.apply(filtered_data)
no_heartbeat.plot()
```

<h2>1.3 Events </h2>

```python

# find the number of event types
events = mne.find_events(no_heartbeat)
#print(events)

#find the number of events
event_count = {}
for i in events:
    event_id = int(i[2])
    if event_id not in event_count.keys():
        event_count[event_id] = 1
    else:
        event_count[event_id] +=1

#print result
for key, value in event_count.items():
    print(f"{key}: {value}")
```

```python
# plot the events using built in function found in documentation at https://mne.tools/stable/generated/mne.viz.plot_events.html#mne.viz.plot_events
fig, ax = plt.subplots(figsize=(20, 6))
plot = mne.viz.plot_events(events, sfreq=no_heartbeat.info['sfreq'], axes =ax)

```

<h3>Are events of the same type clustered in time? Is the sequence of events random or predictable? Could this distribution in time affect brain response? </h3>
The visual/auditory events are clustered toghether, while the smily and button press are at set intervals. The auditory and visual events could potentially be random, but the smiley and button press are predicatable at specifically spaced out intervals.
<br></Br>
My assumption is that we would see changes in brain response at the instances the auditory and Visual events pause. We should also potentially see changes in brain response at the instances the patient presses the button or smiles.

<h2>1.4 Analyze evoked responses </h2>

```python


# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=1, tmin=-0.2, tmax=0.7, picks=['mag'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
left_ear_graph = evoked.plot_joint()
```

<h3>Focus on the brain repsonse within 150 ms of the stimulus. Are the peaks resulting fromt the automatic RMS selection informative?</h3>
We can see that from about 75 ms to 150 ms after the event, we have a spike in brain activity. All the graphed sensors exhibit relatively the same shape (spike at that specific timepoint) so we can conclude that yes, the RMS selection is informative.

<h3>Can you think of a better approach than the rms to select informative peaks? Can you find and visualize a more informative sequence of peaks?
From this, what can you find out about the sequence of events after left ear stimulation</h3>
The RMS basically averages sums up all the electrodes, and then chooses the points with the highest total spike. While this is good, there is potential for an outlier to skew the result. An alternative way that we could potentially select points in time, is by counting the number of signals that exceed a certain threshold, and then select points in time where this count is highest. That way, we would select peaks based on when most of the brain is active, instead of the highest total magnetic field strength.  

When the left ear is stimulated, it looks like 100 ms after the stimulation, the magnetometers read strong electric fields, which means that different Neurons are either depolarizing or hyperpolarizing and thus causing current flow. After the initial spikes, the neurons move toward rest potential, which lessens the current and in turn the magnetic fields.

<h3> Do the same for right ear stimulation, and compare the two response profiles <\h3>

```python
# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=2, tmin=-0.2, tmax=0.7, picks=['mag'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
right_ear_graph = evoked.plot_joint()
```

Looking at the graph, we see the right ear stimulation is very similar to the left. Both spiked at 100 ms and the areas in which that occured are the same as with the left ear stimulation.

<h3> Do the same for visual stimulation. Compare visual to auditory responses. In what respects to the differ? In what repects are they similar?</h3>

```python
# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=3, tmin=-0.2, tmax=0.7, picks=['mag'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
left_eye_graph = evoked.plot_joint()
# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=4, tmin=-0.2, tmax=0.7, picks=['mag'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
left_ear_visual = evoked.plot_joint()
```

Looking at the above visual graphs, we see the major similarity being the spike at 100 ms. However, there are a couple differences we can point out. The first is that the brain seems to process visual stimuli for a longer period of time, as well as relaying information to many different areas of the brain, whereas in the auditory stimulations, the regions of the brain that were active was more concentrated in one area.

Secondly, for the visual, it looks like the magnetic fields have opposite directions, which means that the current is moving in different directions depending on which eye is stimulated. This is in contrast to the auditory stimulations, where both left and right had the magnetic fields move in the same direction.

<h3>Do the same analysis, but for EEG data. Would this lead to the same conclusion?<\h3>

```python
# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=1, tmin=-0.2, tmax=0.7, picks=['eeg'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
left_ear_graph = evoked.plot_joint()

# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=2, tmin=-0.2, tmax=0.7, picks=['eeg'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
right_ear_graph = evoked.plot_joint()

# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=3, tmin=-0.2, tmax=0.7, picks=['eeg'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
left_eye_graph = evoked.plot_joint()

# Epoch the data (200 ms before to 700 ms after the event) (https://mne.tools/stable/generated/mne.Epochs.html#mne.Epochs)
epochs = mne.Epochs(no_heartbeat, events, event_id=4, tmin=-0.2, tmax=0.7, picks=['eeg'])
#epochs.plot()
evoked = epochs.average()
# Visualize the time course of the evoked response
right_eye_graph = evoked.plot_joint()
```

While the overall trend is not what I expected, where the overall brain looks much more active in the EEG's than the MEG's, the thing that was consistent was that the the auditory left and right responses are very similar to each other, but the right and left visual responses differ heavily.
The auditory stimulations;right and left; had the same parts of the brain fire, and current moved in the same direction. However unlike auditory, the right and left visual stimulations caused different parts of the brain to be active depending on which eye recieved stimuli.
