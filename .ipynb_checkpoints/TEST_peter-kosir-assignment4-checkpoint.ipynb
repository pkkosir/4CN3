```python
import mne
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import os
import pathlib
import warnings
import eelbrain
import eelbrain.datasets._alice
from scipy.stats import ttest_1samp #import 1 sample t-test
from scipy.stats import pearsonr #import pearson correlation library 

DATA_ROOT = eelbrain.datasets._alice.get_alice_path() #gets dataset from C:\User\<USERNAME>\Data\Alice

STIMULUS_DIR = DATA_ROOT / 'stimuli' #stimulus data
EEG_DIR = DATA_ROOT / 'eeg' #eeg data

#definitions to be used throughout
LOW_FREQ = 0.5 
HIGH_FREQ = 20 #already have a 20Hz low-pass filter, no change from 3 here

mne.set_log_level('WARNING')  # supresses most useless outputs

print(f"{mne.__version__=}")
print(DATA_ROOT)
```

<strong><font size="6">1</font></strong>

```python
#----------------------# SETUP #----------------------#

SUBJECTS = [path.name for path in EEG_DIR.iterdir() if path.name.startswith('S')]

latencies = [(0,100), (100,200), (200, 300), (300,400), (400,500), (500,600), (900,1000), (1900,2000)] #defines the 6 latency windows btwn 0 and 600ms

# Suppress warning on low pass filter
warnings.filterwarnings("ignore", category=RuntimeWarning, message="The measurement information indicates a low-pass frequency of 20.0 Hz.*")

#----------------------# GENERATE ENVELOPE #----------------------#
envelopes = {}
for stimulus_id in range(1, 13):
    wav = eelbrain.load.wav(STIMULUS_DIR / f'{stimulus_id}.wav')
    envelope = wav.envelope() #computes acoustic envelope
    envelope = eelbrain.filter_data(envelope, LOW_FREQ, HIGH_FREQ, pad='reflect')
    envelope = eelbrain.resample(envelope, 50)
    envelopes[f'{stimulus_id}'] = envelope
```

```python
#----------------------# TRF #----------------------#

decoders = {}
rVals = []

for subject in SUBJECTS:

    #----------# PREPROCESSING #----------#
    raw = mne.io.read_raw(EEG_DIR / subject / f'{subject}_alice-raw.fif', preload = True) #reads in the subjects .fif file 
    raw.filter(LOW_FREQ, HIGH_FREQ, n_jobs = 1) #filters the raw based on tutorial recommendation, defaults to single job running only
    raw.interpolate_bads() #interpolate bad channels

    #----------# EEG PROCESSING #----------#
    events = eelbrain.load.mne.events(raw) #extracts events from the raw data
    
    events['envelope'] = [envelopes[stimulus_id] for stimulus_id in events['event']] #adds the envelopes to the events table
    events['duration'] = eelbrain.Var([envelope.time.tstop for envelope in events['envelope']]) #extracts stimulus duration
    events['eeg'] = eelbrain.load.mne.variable_length_epochs(events, 0, tstop='duration', decim=10, connectivity='auto') #extracts EEG data @ time of envelope

    #----------# TRF PROCESSING #----------#

    decoders[subject] = {}
    rVal = []
    for tmin, tmax in latencies[:]:
        
        trf = eelbrain.boosting('envelope', 'eeg', -tmax/1000, -tmin/1000, delta=0.05, data=events, partitions=4) #converts time into s from ms, and takes 4 partitions (default to 10 which breaks eelbrain)
        decoders[subject][tmin] = trf

        rVal.append(decoders[subject][tmin].r)
    rVals.append(rVal)
```

```python
#----------------------# T-TEST #----------------------#
   
data = np.array(rVals)  

ttest = ttest_1samp(data, 0) #single-sample t-test from the accuracy of each latency window. null-hypothesis is 0, since any value above 0 for "proportion_explained" means some amount of variance/predicatbility
print(ttest)
'''

for tmin, tmax in latencies[:]:
    print(tmin/1000, tmax/1000)
print(latencies)
rows = []
for subject in SUBJECTS:
    row = []
    for tstart, tstop in latencies:
        row.append(decoders[subject][tstart].r)
    rows.append(row)

# Subject x latency matrix
data = np.array(rows)
#----------------------# PLOTTING #----------------------#
'''
x_labels = [f"{tmin}{tmax} ms" for tmin, tmax in latencies] #labels indicating range (for clarity instead of 0ms, 100ms, etc)
y_labels = data.mean(0)
error = ttest.confidence_interval().high - y_labels

plt.figure(figsize=(10,5)) #strched for better readability
plt.errorbar(x = x_labels, y=y_labels, yerr=error, fmt='.-', ecolor='r', capsize=2, label="Decoding Accuracy") #plots decoding, adds red error bars w/ endcaps for readability
plt.xlabel('Latency Windows(ms)')
plt.ylabel('Decoding Accuracy, r')
plt.title('Decoding Accuracy vs Latency')
plt.grid() #shows gridlines for readability
plt.show()
'''

x = [(tstart + tstop) / 2 for tstart, tstop in latencies]
y = data.mean(0)
ci = ttest.confidence_interval()
yerr = ci.high - y


plt.errorbar(x, y, yerr=yerr)
plt.xticks(np.arange(0, 2.1, 0.200))
plt.grid(axis='x')
plt.ylabel('r')
plt.xlabel('Lag window center (s)')
plt.ylim(0, 0.13)
plt.xlim(latencies[0][0], latencies[-1][1])
plt.title("Envelope reconstruction as a function of sound/EEG lag")
'''
```
